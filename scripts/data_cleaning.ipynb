```python
import pandas as pd
import numpy as np

# Load your datasets
train = pd.read_csv(r"C:\Users\adars\OneDrive\Desktop\powerbi\Airline-project\data\raw data\train.csv")
test = pd.read_csv(r"C:\Users\adars\OneDrive\Desktop\powerbi\Airline-project\data\raw data\test.csv")

# Check the dataset sizes
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
print(train.shape)
print(test.shape)
```

```python
df = pd.concat([train, test], ignore_index=True)

print(df.shape)
```

```python
print(df.head())

print(df.info())

print(df.describe())
```

```python
print(df.columns.tolist())
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
print(df.shape)
```

```python
missing = df.isnull().sum()

print(missing[missing > 0])
```

```python
df["Arrival Delay in Minutes"] = (
    df["Arrival Delay in Minutes"]
    .fillna(
        df["Arrival Delay in Minutes"].median()
    )
)
```

```python
print(df.isnull().sum().sum())
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
print(df.columns.tolist())
```

```python
service_cols = [
    'inflight_wifi_service',
    'ease_of_online_booking',
    'food_and_drink',
    'online_boarding',
    'seat_comfort',
    'inflight_entertainment',
    'on-board_service',
    'leg_room_service',
    'cleanliness'
]

df["service_score"] = (
    df[service_cols]
    .mean(axis=1)
)
```

```python
df["delay_bucket"] = pd.cut(
    df["arrival_delay_in_minutes"],
    bins=[-1,15,30,60,10000],
    labels=[
        "0-15",
        "16-30",
        "31-60",
        "60+"
    ]
)
```

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,25,40,60,100],
    labels=[
        "Young",
        "Adult",
        "Middle Age",
        "Senior"
    ]
)
```

```python
df.to_csv(
    "C:\\Users\\adars\\OneDrive\\Desktop\\powerbi\\Airline-project\\data\\processed\\cleaned_airline_data.csv",
    index=False
)
```
